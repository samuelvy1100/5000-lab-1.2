# About This Project

This page provides additional information about the Quarto website.  
It demonstrates how Markdown and Python code can be combined in a Jupyter Notebook.


## Background

Quarto integrates reproducible research tools with modern publishing.  
This notebook illustrates how text and computation can live side by side.


### Objectives

- Demonstrate professional formatting with Markdown  
- Show how to embed Python-generated plots  
- Keep content clean, structured, and reproducible  


```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

# Create the plot
plt.figure(figsize=(6,4))
plt.plot(x, y, label="y = sin(x)")
plt.title("Sample Plot")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.legend()
plt.grid(True)
plt.show()
```
